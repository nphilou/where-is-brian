```python
import numpy as np
import pandas
```

```python
english = pandas.read_csv("train_cap2018.csv")

english.head()
```

```python
col = ['sentences', 'level1']
english = english[col]
english.head()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
english.groupby('level1').sentences.count().plot.bar(ylim=0)
plt.show()
```
